**1.Find a Position in a Genome Minimizing the Skew(BA1F)**

```python
def Pos_In_A_Genome(genome):
    mn = 0
    min_list = []
    cur = 0
    
    for index, ch in enumerate(genome):
        if ch == 'C':
            cur -= 1
        elif ch == 'G':
            cur += 1
        
        if cur < mn:
            mn = cur
            min_list = [index+1]
        elif cur == mn:
            min_list.append(index+1)
    return min_list
genome = input()
min_list = Pos_In_A_Genome(genome)

for idx in min_list:
    print(idx, end=" ")
```

**2.Compute the Hamming Distance Between Two Strings(BA1G)**

```python
def hammingDistance(p, q):
    cnt = 0
    for i in range(len(p)):
        if p[i] != q[i]:
            cnt += 1
    return cnt
```

Input for **problem no 2**

```python
p, q = input(), input()
print(hammingDistance(p, q))
```

**3.Find All Approximate Occurrences of a Pattern in a String(BA1H)**

```python
def approx_pattern_match(t, p, d):
    ids = []
    l = len(p)
    
    for i in range(l-1, len(t)):
        if hammingDistance(t[i-l+1:i+1], p) <= d:
            ids.append(i-l+1)

    return ids
```

Input for **problem no 3**

```python
p, t, d = input(), input(), int(input())
for id in approx_pattern_match(t, p, d):
    print(id, end=" ")
```

**4.Find the Most Frequent Words with Mismatches in a String(BA1I)**

```python
import itertools

def most_freq(t, k, d):
    dt = {}
    
    for s in map(''.join, itertools.product('ACTG', repeat=k)):
        dt[s] = 0
    
    for i in range(k-1, len(t)):
        s = t[i-k+1:i+1]
         
        for kk in dt:
            if hammingDistance(kk, s) <= d:
                dt[kk] += 1
    
    mx = max(dt.values())
    ss = []
    for k in dt:
        if dt[k] == mx:
            ss.append(k)
    return ss
```

Input for **problem no 4**

```python
t, k, d = input(), int(input()), int(input())
for p in most_freq(t, k, d):
    print(p, end=" ")
```

**5.Find Frequent Words with Mismatches and Reverse Complements(BA1J)**

```python
import itertools

def reverse_complement(s):
    ss = ""
    for ch in s:
        if ch == 'T':
            ss += 'A'
        elif ch == 'A':
            ss += 'T'
        elif ch == 'G':
            ss += 'C'
        else:
            ss += 'G'
    return ss[::-1]

def most_freq_plus(t, k, d):
    dt = {}
    
    for s in map(''.join, itertools.product('ACTG', repeat=k)):
        dt[s] = 0
    
    for i in range(k-1, len(t)):
        s = t[i-k+1:i+1]
         
        for kk in dt:
            if hammingDistance(kk, s) <= d:
                dt[kk] += 1
    
    mx = 0
    ss = []
    for k in dt:
        rev = reverse_complement(k)
        if dt[k]+dt[rev] > mx:
            mx = dt[k]+dt[rev]
            ss = [k]
        elif dt[k]+dt[rev] == mx:
            ss.append(k)
    print(mx)
    return ss
```

Input for **problem no 5**

```python
t, k, d = input(), int(input()), int(input())
for p in most_freq_plus(t, k, d):
    print(p, end=" ")
```
