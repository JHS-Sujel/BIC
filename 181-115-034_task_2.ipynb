**1.Find a Position in a Genome Minimizing the Skew(BA1F)**

```python
def Pos_In_A_Genome(genome):
    mn = 0
    min_list = []
    cur = 0
    
    for index, ch in enumerate(genome):
        if ch == 'C':
            cur -= 1
        elif ch == 'G':
            cur += 1
        
        if cur < mn:
            mn = cur
            min_list = [index+1]
        elif cur == mn:
            min_list.append(index+1)
    return min_list
genome = input()
min_list = Pos_In_A_Genome(genome)

for idx in min_list:
    print(idx, end=" ")
```

**2.Compute the Hamming Distance Between Two Strings(BA1G)**

```python
def hammingDist(str1, str2):
    i = 0
    count = 0
 
    while(i < len(str1)):
        if(str1[i] != str2[i]):
            count += 1
        i += 1
    return count
  
str1, str2 = input(), input()
 
print(hammingDist(str1, str2))
```

**3.Find All Approximate Occurrences of a Pattern in a String(BA1H)**

```python
def approx_pattern_match(t, p, d):
    ids = []
    l = len(p)
    
    for i in range(l-1, len(t)):
        if hammingDistance(t[i-l+1:i+1], p) <= d:
            ids.append(i-l+1)

    return ids
```

```python
p, t, d = input(), input(), int(input())
for id in approx_pattern_match(t, p, d):
    print(id, end=" ")
```

**4.Find the Most Frequent Words with Mismatches in a String(BA1I)**

```python
from collections import defaultdict

def neighbour(pattern, mismatch, words):

    bases = ['A', 'T', 'G', 'C']
    for i in range(len(pattern)):

        for j in range(len(bases)):

            new_pattern = pattern[:i] + bases[j] + pattern[i + 1:]

            if mismatch <= 1:
                words.add(new_pattern)

            else:
                neighbour(new_pattern, mismatch - 1, words)


def FindMostFrequentWords(text, k, d):
    AllFrequentWords = defaultdict(int)
    for i in range(len(text) - k + 1):

        FrequentWords = set()
        neighbour(text[i:i + k], d, FrequentWords)
        for words in FrequentWords:
            AllFrequentWords[words] += 1

    for t in AllFrequentWords.keys():
        if AllFrequentWords[t] == max(AllFrequentWords.values()):
            print(t, end=" ")


text = input()
ra = input().split()
k = int((ra[0]))
d = int(ra[1])
FindMostFrequentWords(text, k, d)
```

**5.Find Frequent Words with Mismatches and Reverse Complements(BA1J)**

```python
from collections import defaultdict


def HammingDistance(s1, s2):
    d = sum([1 for i in range(len(s1)) if s1[i] != s2[i]])
    return d


def comp(seq):
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
    bases = list(seq)
    bases = [complement[base] for base in bases]
    return ''.join(bases)


def ReversePattern(s):
    return comp(s[::-1])


def neig(pattern, mismatch, words):


    if mismatch == 0:
        words.add(pattern)



    else:
        bases = ['A', 'T', 'C', 'G']
        for i in range(len(pattern)):
            for j in range(len(bases)):
                new_pattern = pattern[:i] + bases[j] + pattern[i + 1:]
                if mismatch <= 1:
                    words.add(new_pattern)
                else:
                    neig(new_pattern, mismatch - 1, words)


def MostFrequentPattern(text, k, d):
    allfrequentwords = defaultdict(int)
    for i in range(len(text) - k + 1):
        frequentwords = set()
        neig(text[i:i + k], d, frequentwords)

        for words in frequentwords:
            allfrequentwords[words] += 1

    for t in allfrequentwords.keys():
        reverse_k = ReversePattern(t)
        for i in range(len(text) - k + 1):
            if HammingDistance(text[i:i + k], reverse_k) <= d:
                allfrequentwords[t] += 1

    result = set()
    for t in allfrequentwords.keys():
        if allfrequentwords[t] == max(allfrequentwords.values()):
            result.add(t)
            result.add(ReversePattern(t))


    for i in result:
        print(i, end=" ")


text = input()
ra = input().split()
k = int((ra[0]))
d = int(ra[1])
MostFrequentPattern(text, k, d)
```
