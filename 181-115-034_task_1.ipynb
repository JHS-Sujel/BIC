**1.Compute the Number of Times a Pattern Appears in a Text (BA1A)**

```python
def Count(dna, pattern,n,m):
    counter = 0
    for i in range(n):
        if i + m <= n:
            sub = (dna[i: i + m])
            if (sub == pattern):
                counter+= 1
    return counter

dna = input()
pattern = input()
print(Count(dna, pattern,len(dna),len(pattern)))
```

**2.Find the Most Frequent Words in a String (BA1B)**

```python
from collections import defaultdict

def MostFrequentK_mer(dna, k):
    l = len(dna)
    most_kmer_count = defaultdict(int)

    for i in range(l - k + 1):
        most_kmer_count[dna[i:i+k]] += 1

    temp = max(most_kmer_count.values())

    return ' '.join([k for k, v in most_kmer_count.items() if v == temp])

if __name__ == '__main__':
    dna = input()
    k   = int(input())
    print(MostFrequentK_mer(dna, k))
```

**3.Find the Reverse Complement of a String (BA1C)**

```python
def cmp(p):

    ans = ""

    n=len(p)

    for i in range (n):
        if p[i] == 'A':
            ans += 'T'
        elif p[i] == 'T':
            ans += 'A'

        elif p[i] == 'C':
            ans += 'G'

        else:
            ans += 'C'

    Ans= ans[::-1];

    return Ans;

p=input()
print(cmp(p))
```

**5.Find Patterns Forming Clumps in a String (BA1E)**

```python
def PatternsFormingClumps(gene, L, t, k):
    window = ""
    clumps = set()
    map_kmer_idx = {}
    
    for index, ch in enumerate(gene):
        window += ch
        if index >= k:
            window = window[1:]
                        
            if window not in map_kmer_idx:
                map_kmer_idx[window] = []
            
            r = len(map_kmer_idx[window])
            
            if r >= t:
                map_kmer_idx[window].pop(0)
            map_kmer_idx[window].append(index)

            if r == t-1 and map_kmer_idx[window][r] - map_kmer_idx[window][0] + 1 <= L:
                clumps.add(window)

    return clumps
gene = input()
k, L, t = [int(x) for x in input().split()]

for k_mer in PatternsFormingClumps(gene, L, t, k):
    print(k_mer, end=" ")
```
